```python
import uproot
import numpy as np

signal_data_file = "../mu3e_root_data/run_sig_simi-sort.root"
with uproot.open(signal_data_file) as file:
    event_data = file["mu3e"].arrays(library="pd")
    mc_hit_truth_time = file["mu3e_mchits"]["time"].arrays(library="pd")["time"]

mppc_frame_length = (
    event_data["fibremppc_mc_i"]
    .map(
        lambda ids: (
            mc_hit_truth_time.iloc[ids].max() - mc_hit_truth_time.iloc[ids].min()
            if len(ids) > 0
            else np.nan
        )
    )
    .dropna()
    .to_numpy()
)
np.max(mppc_frame_length), np.mean(mppc_frame_length), np.median(mppc_frame_length)
```
